# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in a wide range of formats and styles. It provides a flexible and comprehensive toolkit for generating various types of plots, charts, and graphs. Matplotlib is widely used in data analysis, scientific research, and data visualization tasks due to its ease of use and extensive customization options.

Here are five types of plots that can be created using the pyplot module of Matplotlib:

Line Plot: A line plot displays data points connected by straight lines, commonly used to visualize trends or patterns in data over a continuous axis.

Bar Plot: A bar plot represents categorical data with rectangular bars, where the length or height of each bar corresponds to the value of the data it represents. It's often used to compare values across different categories.

Scatter Plot: A scatter plot displays individual data points as dots on a two-dimensional plane, with one variable plotted on the x-axis and another on the y-axis. It's used to observe the relationship between two continuous variables.

Histogram: A histogram is used to visualize the distribution of a single continuous variable. It groups data into bins and represents the frequency or count of data points within each bin as bars.

Pie Chart: A pie chart displays data as slices of a circular pie, where each slice represents a portion of the whole. It's commonly used to show the proportion of different categories within a dataset.

A scatter plot is a type of data visualization that uses points (or dots) to represent individual data observations. It's particularly useful for showing the relationship between two continuous variables. In a scatter plot, each data point is plotted on a two-dimensional plane, with one variable plotted along the x-axis and the other variable plotted along the y-axis.

The main purpose of a scatter plot is to visualize the distribution and correlation between the two variables. It helps you understand how changes in one variable might be related to changes in the other variable. Scatter plots are particularly effective when dealing with large datasets because they allow you to quickly identify patterns, trends, clusters, and outliers.

Here are some key characteristics of scatter plots:

Individual Data Points: Each data observation is represented as a single point on the plot.

Axes: The x-axis and y-axis represent the two variables being compared. Each data point's position is determined by its values along these axes.

Correlation: The pattern formed by the points on the plot can provide insight into the correlation between the two variables. Positive correlation implies that as one variable increases, the other tends to increase as well. Negative correlation implies that as one variable increases, the other tends to decrease.

Trends and Patterns: Scatter plots help identify trends, patterns, or clusters of data points. These patterns can give valuable insights into the data.

Outliers: Outliers, which are data points that deviate significantly from the general trend, can also be identified in scatter plots.

```python
import matplotlib.pyplot as plt
import numpy as np 
np.random.seed(3) 
x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("Scatter Plot")
plt.show()
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. This is particularly useful when you want to display multiple plots side by side or in a grid arrangement within a single figure window. Each subplot can represent a different plot type, data comparison, or aspect of the data.

The subplot() function takes three arguments: the number of rows in the grid, the number of columns in the grid, and the index of the current subplot. The index increases from left to right and top to bottom, starting at 1.

Here are some common use cases for using the subplot() function:

Comparing Multiple Plots: You might want to compare different visualizations or representations of the same data. For example, you could create side-by-side line plots with different settings to compare the effects of varying parameters.

Multi-View Data Analysis: If you're analyzing a dataset, you might want to create multiple plots that provide different perspectives on the data. Using subplots, you can view histograms, scatter plots, and other visualizations simultaneously.

Data Exploration: When exploring a dataset, you might want to visualize different subsets of the data in separate subplots. This can help you quickly identify patterns, outliers, and relationships.

Dashboard or Report Creation: If you're creating a dashboard or a report with multiple visualizations, subplots allow you to arrange them neatly within a single figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='*')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='v')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='o')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='+')
plt.title('Line Plot 4')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 


A bar plot is a type of data visualization that uses rectangular bars to represent and compare the values of different categories or groups. It's a simple yet effective way to display categorical data and their corresponding quantities. Each bar's length or height is proportional to the value it represents, allowing for easy visual comparison between categories.

Bar plots are widely used for various purposes, including:

Comparison of Categorical Data: Bar plots are particularly useful for comparing the frequency, count, or values of different categories. For example, you might use a bar plot to compare the sales performance of different products, the distribution of votes for different political parties, or the average scores of students from different schools.

Visualizing Distribution: Bar plots can be used to visualize the distribution of categorical data. For instance, you could create a histogram-like bar plot to show the distribution of ages in different age groups.

Showing Trends Over Time: When combined with a time component, bar plots can be used to show how values in different categories change over time. This is often called a "grouped bar chart," where each category has bars representing its values at different time points.

Comparing Subcategories: Stacked bar plots allow you to compare subcategories within each main category. This is useful when you want to visualize how different components contribute to a whole.

Highlighting Discrepancies: Bar plots can help highlight discrepancies or differences between categories, making it easy to identify outliers or areas of interest.

In a bar plot, the x-axis represents the categories or groups being compared, while the y-axis represents the values associated with those categories. The bars can be oriented either horizontally (barh) or vertically (bar), depending on your preference and the space available.

```python
import numpy as np 
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"]) 
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot. box1 = np.random.normal(100, 10, 200) box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution and summary statistics of a dataset. It displays the median, quartiles, and potential outliers in a compact manner, making it an effective tool for understanding the spread and central tendency of data. Box plots are especially useful for comparing distributions across different groups or categories.

A box plot consists of several key elements:

Box: The box in the plot represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile). The height of the box indicates the spread of the middle 50% of the data.

Whiskers: The whiskers extend from the box to the minimum and maximum values within a certain distance from the quartiles. They provide information about the spread of the entire dataset, except for potential outliers.

Median Line: A vertical line within the box represents the median, which is the value that separates the lower and upper halves of the dataset.

Outliers: Data points that fall outside the whiskers are considered outliers and are usually plotted as individual points.

Box plots are used for various purposes, including:

Visualizing Distribution: Box plots provide a clear visualization of the data distribution, indicating skewness and the presence of potential outliers.

Comparing Groups: When you have multiple groups or categories of data, box plots help compare their distributions. This can be particularly useful for identifying differences and similarities between groups.

Identifying Outliers: Box plots highlight potential outliers, which are data points that are significantly different from the rest of the data. These outliers can be important for data quality assessment.

Summarizing Statistics: Box plots present key summary statistics such as the median, quartiles, and potential range of the data in a concise manner.

Data Exploration: Box plots are often used as a first step in data exploration, helping analysts quickly grasp the nature of the data.

```python
import numpy as np 
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Boxes')
plt.ylabel('Values')
plt.title('Box Plot Comparison')
plt.show()
```
